```python
def interm_pivot(columns):
    for col in columns:
        df_A[col].apply(lambda x:str(x))
        df_B[col].apply(lambda x:str(x))
    df_A["pivot"] = df_A[columns].apply(lambda x: ''.join(x), axis = 1)
    df_B["pivot"] = df_B[columns].apply(lambda x: ''.join(x), axis = 1) 
    return df_A,df_B
```

```python
def interm_lev(columns):
    for col in columns:
        df_A[col].apply(lambda x:str(x))
        df_B[col].apply(lambda x:str(x))
        # Conversion de la colonne 'colonne_float' en string
        df_A['tx_amount'] = df_A['tx_amount'].astype(str)
        df_B['tx_amount'] = df_B['tx_amount'].astype(str)
    df_A["lev"] = df_A[columns].apply(lambda x: ''.join(x), axis = 1)
    df_B["lev"] = df_B[columns].apply(lambda x: ''.join(x), axis = 1) 
    return df_A,df_B
```

```python
def poolbis(df1,df2,columns,d,discr):
    df1,df2 = interm_pivot(columns)
    df2["distlev"]=d+1
    for i in range(len(df1)):
        if i%1==0:
            print(i)
        for j in range(len(df2)):
            df2["distlev"]=Levenshtein.distance(df1["pivot"][i], df2["pivot"][j])
            if j%100000==0:
                print(j)
        dfB_sub = df2[df2["distlev"] <= d]
        print(dfB_sub)
        for j in range(len(df2)):  
            lev = []
            for k in range(len(dfB_sub)): 
                dist_lev = Levenshtein.distance(df1["lev"][i], df2["lev"][k])
                lev.append(dist_lev)
                if min(dist_lev) >= discr:
                    id_ = dfB_sub["index"][np.argmin(lev)]
                    df1["matching"][i] = [df1["index"][i],id_]
    return df1   
```

```python
def pool(df1,df2,columns,discr):
    df1,df2 = interm_pivot(columns)
    for i in range(len(df1)):
        dfB_sub = df2[df2["pivot"] == df1["pivot"][i]]
        if i%1000==0:
            print(i)  
        for k in range(len(dfB_sub)): 
            d = discr+1
            idm = 0
            dist_lev = Levenshtein.distance(df1["lev"][i], df2["lev"][k])
            if dist_lev < d:
                idm=dfB_sub["index"][k]
                d = dist_lev
        if idm <= discr:
            df1["matching"][i] = idm  
    return df1  
```

```python
def precision(df1,df2):
    count = 0
    for i in range(len(df1)):
        if df1['tx_id'][i] == df2['tx_id'][df1['matching'][i]]:
            count += 1
    return count/len(df1)
```
